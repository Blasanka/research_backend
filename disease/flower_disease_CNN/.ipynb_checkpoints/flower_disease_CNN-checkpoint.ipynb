```python
two = []
for i, file in enumerate(os.listdir('DATASET/Training Samples/2/')):
    if i < 100:
        image = Image.open('DATASET/Training Samples/2/'+file)
        image = np.asarray(image)
        two.append(rgb2gray(image))
two = np.asarray(two)
x_train.append(two)
```

```python
data = np.asarray(x_train)
```

```python
data.shape
```

```python
x_train = data.reshape((400, 64, 64,1))
```

```python
x_train.shape
```

```python
y = [0 for i in range(100)]
```

```python
y.extend([1 for i in range(100)])
```

```python
y.extend([2 for i in range(100)])
```

```python
y.extend([3 for i in range(100)])
```

```python
# y.extend([4 for i in range(100)])
```

```python
y= to_categorical(y)
```

```python
y_train = y
```

```python
y_train
```

```python
model = keras.Sequential()
```

```python
model.add(keras.layers.Conv2D(64, (3, 3), activation = 'relu',
                              input_shape = (64, 64, 1)))
```

```python
model.add(keras.layers.MaxPool2D((2, 2)))
```

```python
model.add(keras.layers.Conv2D(64, (3, 3), activation = 'relu'))
```

```python
model.add(keras.layers.MaxPool2D((2, 2)))
```

```python
model.add(keras.layers.Conv2D(64, (3, 3), activation = 'relu'))
```

```python
model.add(keras.layers.MaxPool2D((2, 2)))
```

```python
model.add(keras.layers.Conv2D(64, (3, 3), activation = 'relu'))
```

```python
model.add(keras.layers.Flatten())
```

```python
model.add(keras.layers.Dense(200, activation = 'relu'))
```

```python
model.add(keras.layers.Dense(100, activation = 'relu'))
```

```python
model.add(keras.layers.Dense(4, activation = 'softmax'))
```

```python
model.summary()
```

```python
model.compile(optimizer = 'Adam', loss = 'categorical_crossentropy',
              metrics = ['accuracy'])
```

```python
model.fit(x_train, y_train, epochs = 50, batch_size = 64)
```
